```python
#Data Analysis

from datasets import load_dataset
```

```python
import pandas as pd
import numpy as np
```

```python
#dataset for pandas pratice

gion = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/noc_regions.csv')
bios = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv')
coffee = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/warmup-data/coffee.csv')
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
#cleanup job_posted_date to date
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

#df.info() # verifiy that this is done
```

```python
#add new columns based on other columns
df['is_data_analyst'] = (df['job_title_short'] == 'Data Analyst').astype(int)
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
#sort the data frame
df.sort_values(by='job_posted_date', inplace= True)
```

```python
#drop unnecessary columns
df.drop('salary_hour_avg', axis=1, inplace=True)

```

```python
#remove null or NaN values
df.dropna(axis=0, subset=['salary_year_avg'], inplace=True)
```

```python
#analyse the dataset
salary_min_line = df['salary_year_avg'].idxmin()
```

```python
#group data
df.groupby(by=['job_title_short', 'job_country'])['salary_year_avg'].agg(['min', 'max', 'median'])
```

```python
#analyse job based on USA only
us_jobs = df[df['job_country'] == 'United States']
```

```python
us_jobs.groupby('job_title_short')[['salary_year_avg']].agg(['min', 'max', 'median'])
```
